```python
from math import *
import seaborn as sns
import numpy as np
```

${ (1+r)^n * r = 1 }$

$(1+r)^n = r^{-1}$

$log_{1+r}{(1+r)^n} = log_{1+r}{r^{-1}}$

$n = -log_{1+r}{r}$

```python
def turns(r): return -log(r)/log(1+r)
```

```python
turns(0.025)
```

<p>
    ${ (1+r)^n * r = 1 }$
</p>
<p>
    ${ 1+r = r^{-n} }$
</p>
<p>
    ${ \frac{1}{1+r} = r^n }$
</p>


```python
def term(r,n):  return (1+r)**n * r - 1
def solve(n):
    a = 0.01
    b = 0.2

    mid = (a+b)/2
    val = term(mid, n)
    while abs(val) > 0.001:
        if val < 0: a = mid
        else: b = mid
        mid = (a+b)/2
        val = term(mid, 20)
    return mid
solve(20), turns(0.10)
```
